```python
#Doughnut Chart
import numpy as np
from matplotlib import pyplot as plt

tech = ['Google','Apple','Amazon','Tesla','Microsoft','OpenAI','Netflix']
quantity = [31,24,19,11,6,10,9]
plt.pie(quantity,labels= tech,autopct= '%0.1f%%',radius= 3)
plt.pie([3],colors= ['indigo'],radius=2)
plt.show()
```

```python
plt.pie(quantity,labels= tech,autopct= '%0.1f%%',radius= 3)
plt.pie([1],colors= ['white'],radius=2)
plt.show()
```
